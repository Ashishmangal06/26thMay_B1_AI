```python
#import Modules
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df_train = pd.concat([df_train, pd.get_dummies(df_train['Type'])],axis=1)
df_train.drop('Type',axis=1,inplace=True)
df_train.head()
```

```python
df_train = pd.concat([df_train, pd.get_dummies(df_train['City Group'])],axis=1)
df_train.drop('City Group',axis=1,inplace=True)
df_train.head()
```

```python
df_train = pd.concat([df_train, pd.get_dummies(df_train['City'])],axis=1)
df_train.drop('City',axis=1,inplace=True)
df_train.head()
```

```python
df_train['Open Date'][0].split('/')
```

```python
df_train['Date'] = df_train['Open Date'].apply(lambda x: int(x.split('/')[0]))
```

```python
df_train['Month'] = df_train['Open Date'].apply(lambda x: int(x.split('/')[1]))
df_train['Year'] = df_train['Open Date'].apply(lambda x: int(x.split('/')[2]))
df_train.head()
```

```python
df_train.drop('Open Date',axis=1,inplace=True)
df_train.head()
```

```python
df_train.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
X = df_train.loc[:, df_train.columns != 'revenue']
y = df_train.loc[:, df_train.columns == 'revenue']
tree = DecisionTreeRegressor().fit(X, y)
```

```python
import numpy as np
d = {'Importance': np.round(tree.feature_importances_, 3) , 'Features': X.columns}
feature_imp = pd.DataFrame(data=d)
```

```python
feature_imp
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
X_main = df_train[feature_imp.Features.values]
y_main = df_train['revenue']
scaler_train = StandardScaler()
scaler_train.fit(X_main)
X_main_scaled = scaler_train.transform(X_main)

# Splitting the data into training and testing data
linear_regr = LinearRegression()
linear_regr.fit(X_main_scaled, y)
y_pred = linear_regr.predict(X_main_scaled)
accuracy = linear_regr.score(X_main_scaled,y)
print("Train Accuracy {}%".format(int(round(accuracy *100))))
print("Training RMSE Linear regression ",mean_squared_error(y, y_pred, squared=False))
```

```python
feature_imp_order = feature_imp.sort_values('Importance', ascending=False).reset_index(drop=True)[0:20]
feature_imp_order
```

```python
final_model = linear_regr
pd.to_pickle(final_model, 'RestaurantRevenuePrediction')
```
